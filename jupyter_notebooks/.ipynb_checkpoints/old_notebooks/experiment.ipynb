### Correct for h

```python
Tf_fan_model = Tf_fan.to_numpy()
Tb_fan_model = Tb_fan.to_numpy()
time_fan_model = time_fan.to_numpy()
h_f = dft.natural_convection(Tf_fan_model)
h_b = dft.natural_convection(Tb_fan_model)
C = 0.65
n = 0.25
h_f.custom(C, n)
h_b.custom(C, n)

def q(x):
    goal = sensor_1_no_fan.q_inc
    to_correct = dft.one_dim_conduction(Tf_fan_model, Tb_fan_model, time_fan_model, x, h_b.h, model='one_d_conduction').q_inc
    mse = mean_squared_error(goal, to_correct)
    return mse
   
```

```python
result = opt.minimize(q, h_f.h, method='Powell',options={'xatol': 1, 'disp': True})
```

```python
plt.plot(time_fan_model,result['x'], label='h front corrected', color='red',lw=3)
plt.plot(time_fan_model,h_f.h, label='h original',color='dodgerblue',lw=3)
plt.plot(time_fan_model,savgol_filter(result['x'], 101, 5), lw=3, label='smoothed', color='navy',linestyle='--')
plt.xlabel('Time (s)')
plt.ylabel('h (W/m$^2$K)')
plt.xlim(250,550)
plt.ylim(0,60)
plt.legend();
```

```python
corrected = dft.one_dim_conduction(Tf_fan_model, Tb_fan_model, time_fan_model, result['x'], h_b.h, model='one_d_conduction').q_inc
```

```python
plt.plot(time_fan_model,corrected, label='corrected', color='navy', lw=2)
plt.plot(time_fan_model,sensor_1_no_fan.q_inc, label='actual', color='red', lw=2)
plt.plot(time_fan_model,sensor_1_fan.q_inc, label='original', color='dodgerblue', lw=2)
plt.title('Corrected heat flux for fan')
plt.xlabel('Time (s)')
plt.ylabel('Heat flux kW/m$^2$')

plt.legend();
```

# Find h from additional TC??

$$ \rho c V \frac{dT}{dt}= \alpha A q''_{inc} - hA(T-T_{\infty}) - q^{''}_{rad}A$$

$$ h= \frac{1}{A(T-T_{\infty})} \left[-\rho c V \frac{dT}{dt}+ \alpha A q''_{inc}  - q^{''}_{rad}A\right] $$

$$q^{''}_{rad}=\frac{J_{TC} - E_{b,TC}}{\frac{1-\epsilon}{\epsilon}}$$

Need to find $J_{TC}$ from circuit

$$[A][J]=[B]$$

$$[J]=[A]^{-1}[B]$$

```python
T_inf = 21
D_TC = 1.4e-3
rho_TC = 8665
C_TC = 485.6
k_TC = 24.4
A_TC = 4*np.pi*(D_TC/2)**2
V_TC = (4/3)*np.pi*(D_TC/2)**3
eps_TC = 0.7
alpha_TC = 0.7
```

```python
DFT_area = 0.0762**2
eq_r_DFT = np.sqrt(DFT_area/np.pi)
h_23 = 20e-2 + D_TC/2 # height from DFT to center of sphere
cone_area = (10e-2)**2
eq_r_cone = np.sqrt(cone_area/np.pi)
h_21 = 20e-2 + D_TC/2 # height from cone to center of sphere
```

```python
def F_ab(r_a, r_b, h): # a is the large one
    R_i = r_a/h
    R_j = r_b/h
    S = 1 + (1+R_j**2)/R_i**2
    return 0.5*(S-(S**2 - 4*(r_b/r_a)**2)**0.5) 
```

```python
F_ab(eq_r_cone, eq_r_DFT, 40e-2)
```

```python
# Sphere to disk (model DFT as disk)

A_1 = DFT_area
A_2 = A_TC
A_3 = cone_area
A_4 = 2*eq_r_cone*np.pi*40e-2 + (np.pi*eq_r_cone**2 - DFT_area)

A_23 = eq_r_DFT/h_23
F_23 = 0.5*np.sqrt(1-1/np.sqrt(1+A_23**2))
A_21 = eq_r_cone/h_21
F_21 = 0.5*np.sqrt(1-1/np.sqrt(1+A_21**2))
F_24 = 1 - F_21 - F_23
F_31 = F_ab(cone_area, DFT_area, 40e-2)
F_13 = F_31*A_3/A_1
F_34 = 1-F_31
F_43 = F_34*A_3/A_4
F_41 = 1 - F_43


R_12 = 1/(F_21*A_2)
R_13 = 1/(F_31*A_3)
R_14 = 1/(F_41*A_4)
R_23 = 1/(F_23*A_2)
R_24 = 1/(F_24*A_2)
R_34 = 1/(F_34*A_3)
R_2 = (1-eps_TC)/(eps_TC*A_2)
R_3 = (1-0.94)/(0.94*A_3)
```

```python
def E_b(T):
    return 5.67e-8*T**4

def J_mat(R_12, R_13, R_14, R_23, R_24, R_34, R_2, R_3):
    return np.array([[-1/R_12-1/R_14-1/R_13, 1/R_12, 1/R_13, 1/R_14],[1/R_12, -1/R_24-1/R_12-1/R_2-1/R_23,1/R_23, 1/R_24]
                  ,[1/R_12, 1/R_23, -1/R_24-1/R_13-1/R_23-1/R_3, 1/R_34],[0,0,0,1]])

def B_mat(q_inc, A_cone, T_TC, T_DFT, T_inf):
    EB_TC = E_b(T_TC)
    EB_DFT = E_b(T_TC)
    EB_inf = E_b(T_TC)
    return np.array([q_inc, EB_TC/R_2, EB_DFT/R_3, EB_inf ])
```

```python
T_TC = Tinf_fan.to_numpy()
```

```python
dTdt = np.gradient(T_TC,time_fan_model)
```

```python
h = 1/(A_TC*(T_TC - T_inf)) * (-rho_TC*C_TC*V_TC*dTdt + alpha_TC*A_TC*sensor_1_fan.q_inc*1e3 - eps_TC*5.67*1e-8*A_TC*((T_TC+273)**4-(T_inf+273)**4))
```

```python
plt.plot(time_fan_model,h, lw=3, color='red')
plt.xlabel('Time (s)')
plt.ylabel('h thermocouple (W/m$^2$K)')
plt.title('h for thermocouple')
plt.figure()
plt.plot(time_fan_model,h, lw=3, color='red')
plt.ylim(0,275)
plt.xlim(250,8*60)
plt.xlabel('Time (s)')
plt.ylabel('h thermocouple (W/m$^2$K)')
plt.title('h for thermocouple trimmed');
```

```python
h_trim = h[400:][0:550]
T_TC_trim = T_TC[400:][0:550]
time_trim = time_fan[400:][0:550]
Tf_trim = Tf_fan_model[400:][0:550]
Tb_trim = Tb_fan_model[400:][0:550]
```

```python
R_air = 287.05
P_atm = 101325

def density(P_atm, R_air, T): # T in K
    return P_atm/(R_air*T)
```

```python
def correlation_sphere(D,h,T,T_inf): # T in C
    
    T_eval = T_inf
    nu_inf = dft.air_props(T_eval).kinematic_viscosity(T_eval)
    nu_s = dft.air_props(T).kinematic_viscosity(T)
    alpha = dft.air_props(T_eval).thermal_diffusivity(T_eval)
    k = dft.air_props(T_eval).thermal_conductivity(T_eval)
    Pr = nu_inf/alpha
    Nu = h*D/k
    Re = np.zeros(T.shape[0])
    mu_inf = nu_inf*density(P_atm, R_air,T_inf+273)
    mu_s = nu_s*density(P_atm, R_air, T+273)
    mu_ratio = mu_inf/mu_s

    def f(x): 
        return 0.4*x**0.5 + 0.06*x**(2/3) - (Nu_eq-2)/(Pr**0.4*mu_ratio_eq**0.25)

    
    for idx in range(T.shape[0]):
        Nu_eq = Nu[idx]
        mu_ratio_eq = mu_ratio[idx]
        Re[idx], info = opt.brentq(f,0, 5000, full_output=True)
        # this returns Re(TC) given h TC
        
    u_char = Re*nu_inf/D # get characteristic velocity
    return u_char    
```

```python
u = correlation_sphere(D_TC,h_trim, T_TC_trim, T_inf)
```

```python
plt.plot(time_trim, u, lw=3,color='red')
plt.xlabel('Time (s)')
plt.ylabel('Characteristic velocity (m/s)')
plt.title('characteristic velocity obtained from h for TC');
```

### Check what Re looks like for DFT to pick right correlation

```python
T_eval = 0.5*(Tf_trim + T_inf)
nu = dft.air_props(T_eval).kinematic_viscosity(T_eval)
Res = u*0.0762/nu
np.max(Res)
#mostly laminar
```

```python
def DFT_correlation(u, L_ch, T):
    T_eval = 0.5*(T + T_inf)
    nu = dft.air_props(T_eval).kinematic_viscosity(T_eval)
    alpha = dft.air_props(T_eval).thermal_diffusivity(T_eval)
    k = dft.air_props(T_eval).thermal_conductivity(T_eval)
    Pr = nu/alpha
    Re = u*L_ch/nu
    Nu = 0.664*Re**0.5*Pr**(1/3)
    return k*Nu/L_ch # returns h
```

```python
h_DFT = DFT_correlation(u, 0.0762, Tf_trim)
```

```python
plt.plot(time_trim, h_DFT, color='dodgerblue',lw=3)
plt.xlabel('Time (s)')
plt.ylabel('h (W/m$^2$K)')
plt.title('h for DFT after first iteration');
```

# Automate process

```python
# Initialize
h_f = dft.natural_convection(Tf_fan_model)
h_b = dft.natural_convection(Tb_fan_model)
C = 0.65
n = 0.25
h_f.custom(C, n)
h_b.custom(C, n)
q_inc = dft.one_dim_conduction(Tf_trim, Tb_trim, time_trim.to_numpy(), h_f.h[400:][0:550], h_b.h[400:][0:550], model='one_d_conduction').q_inc
dTdt_trim = dTdt[400:][0:550]

n_iter = 100

mean_qinc = np.zeros(n_iter)
q_inc_idx = np.zeros([n_iter,time_trim.shape[0]])
h_idx = np.zeros([n_iter,time_trim.shape[0]])

for iter_idx in tqdm(range(n_iter)):
    q_inc_idx[iter_idx,:] = q_inc
    mean_qinc[iter_idx] = np.mean(q_inc)
    h_TC = 1/(A_TC*(T_TC_trim - T_inf)) * (-rho_TC*C_TC*V_TC*dTdt_trim + alpha_TC*A_TC*q_inc*1e3 - eps_TC*5.67*1e-8*A_TC*((T_TC_trim+273)**4-(Tf_fan_model[0]+273)**4))
    u = correlation_sphere(D_TC, h_TC, T_TC_trim, T_inf)
    h_DFT = DFT_correlation(u, 0.0762, Tf_trim)
    h_idx[iter_idx,:] = h_DFT
    q_inc = dft.one_dim_conduction(Tf_trim, Tb_trim, time_trim.to_numpy(), h_DFT, h_b.h[400:][0:550], model='one_d_conduction').q_inc
    
    

    
    
```
