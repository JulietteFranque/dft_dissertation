```python
import os
os.chdir('../')
from forced_convection_dft.forward_models import ForwardModelDft, ForwardModelThermocouple
from forced_convection_dft.correlations import PlateCorrelations, SphereCorrelations
from dft import dft_models, heat_transfer_coefficients
os.chdir('jupyter_notebooks')
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 500
time_vector = np.linspace(0, 600, n_points)
velocity =  np.ones(n_points) * 5


diameter_tc = 1.4e-3
density_tc = 8665
heat_capacity_tc = 485.6 
emissivity_tc=0.8
alpha_tc=0.8

dft_thickness =  0.0016
length_insulation = .019
dft_side_length = .0762

temperature_initial = 21 + 273
```

```python
incident_heat_flux = time_vector ** 2 * .000025 * 1e3
#incident_heat_flux = np.ones(n_points) * 10 * 1e3
#incident_heat_flux[time_vector < 120] = 0
#incident_heat_flux[time_vector > 8 * 60] = 0

plt.plot(time_vector, incident_heat_flux / 1e3)
forward_dft = ForwardModelDft(number_of_nodes=10, time_vector=time_vector, insulation_thickness=length_insulation, plate_thickness=dft_thickness, incident_heat_flux=incident_heat_flux,
                 characteristic_velocity=velocity, initial_temperature=temperature_initial, temperature_ambient=temperature_initial, temperature_surroundings=temperature_initial, plate_side_length=dft_side_length)
temperature_front, temperature_back, temperature_insulation, coefficient_front, coefficient_back = forward_dft.run()
forward_tc = ForwardModelThermocouple(time_vector, incident_heat_flux, velocity, temperature_initial,
                 temperature_initial, temperature_initial, emissivity_tc, alpha_tc, diameter_tc, heat_capacity_tc,
                 density_tc)
temperature_tc = forward_tc.run()
```

```python
plt.plot(temperature_tc-273, label='tc')
plt.plot(temperature_front-273, label='front')
plt.plot(temperature_back-273, label='back')
plt.legend()
```

```python
def find_thermocouple_q_inc(h_tc, thermocouple_temperature, ambient_temperature, temperature_surroundings, diameter, emissivity, absorptivity, heat_capacity, density):
    convective_heat_flux = h_tc * (thermocouple_temperature - ambient_temperature)
    area = 4 * np.pi * (diameter / 2) ** 2
    volume = 4 / 3 * np.pi * (diameter / 2) ** 3
    emitted_heat_flux = emissivity * 5.67e-8 * (
            thermocouple_temperature ** 4 - temperature_surroundings ** 4)
    energy_stored = (density * heat_capacity * volume) * np.gradient(thermocouple_temperature, time_vector)
    incident_heat_flux = (energy_stored/area + convective_heat_flux + emitted_heat_flux) / absorptivity
    return incident_heat_flux / 1e3
```

```python
def minimize_error(velocity, ambient_temperature, temperature_tc, temperature_front_dft, temperature_back_dft, dft_side_length, diameter, emissivity, absorptivity, heat_capacity, density):
    h_front_dft = PlateCorrelations(surface_temperature=temperature_front_dft, ambient_temperature=ambient_temperature,characteristic_length=dft_side_length).find_forced_convection_heat_transfer_coefficient(velocity)
    h_back_dft = heat_transfer_coefficients.natural_convection(
            (temperature_back + temperature_initial).reshape(-1, 1) / 2, Kelvin=True,
           T_infty=temperature_initial).custom(0.65, 0.25).flatten()
    h_tc = SphereCorrelations(
                surface_temperature=temperature_tc,
                ambient_temperature=ambient_temperature,
                characteristic_length=diameter).find_forced_convection_heat_transfer_coefficient(velocity)
    q_inc_plate = dft_models.one_dim_conduction(T_f=temperature_front, T_b=temperature_back, time=time_vector, h_f=coefficient_front, h_b=coefficient_back, model='one_d_conduction', Kelvin=True, T_inf=temperature_initial, T_sur=temperature_initial).q_inc
    q_inc_tc = find_thermocouple_q_inc(h_tc, temperature_tc, ambient_temperature, ambient_temperature, diameter, emissivity, absorptivity, heat_capacity, density)
    #plt.plot(q_inc_plate, label='plate')
    #plt.plot(q_inc_tc, label='tc')
    #plt.legend()
    return np.abs(q_inc_plate - q_inc_tc).sum(), q_inc_plate
    
    
```

```python
minimize_error(velocity, temperature_initial, temperature_tc, temperature_front, temperature_back, dft_side_length, diameter_tc, emissivity_tc, alpha_tc, heat_capacity_tc, density_tc);


```

```python
from scipy.optimize import minimize
```

```python
x0 = np.ones(time_vector.shape[0])
res = minimize(lambda x: minimize_error(x, temperature_initial, temperature_tc, temperature_front, temperature_back, dft_thickness, diameter_tc, emissivity_tc, alpha_tc, heat_capacity_tc, density_tc )[0], x0, method='powell',
               options={'ftol': .5, 'disp': True}, bounds=list((((1e-3, np.inf), ) * time_vector.shape[0])))


```

```python
plt.plot(res.x)
plt.ylim(0, 10)
```

```python
q_recov = minimize_error(res.x, temperature_initial, temperature_tc, temperature_front, temperature_back, dft_thickness, diameter_tc, emissivity_tc, alpha_tc, heat_capacity_tc, density_tc )[1]
plt.plot(q_recov)
plt.plot(incident_heat_flux/1e3)
```
