# Find h from additional TC??

$$ \rho c V \frac{dT}{dt}= - hA(T-T_{\infty}) + q^{''}_{rad}A$$

$$ h= \frac{1}{A(T-T_{\infty})} \left[-\rho c V \frac{dT}{dt}+  q^{''}_{rad}A\right] $$

$$q^{''}_{rad}=\frac{J_{TC} - E_{b,TC}}{\frac{1-\epsilon}{\epsilon}}$$

Need to find $J_{TC}$ from circuit

$$[A][J]=[B]$$

$$[J]=[A]^{-1}[B]$$

```python
T_inf = 21
D_TC = 1.4e-3
rho_TC = 8665
C_TC = 485.6
k_TC = 24.4
A_TC = 4*np.pi*(D_TC/2)**2
V_TC = (4/3)*np.pi*(D_TC/2)**3
eps_TC = 0.99
eps_other = .99
alpha_TC = 0.7
```

```python
DFT_area = 0.0762**2
eq_r_DFT = np.sqrt(DFT_area/np.pi)
h_23 = 2e-2 # height from DFT to center of sphere
cone_area = (20e-2)**2
eq_r_cone = np.sqrt(cone_area/np.pi)
total_height = 22e-2
h_21 = total_height - h_23  # height from cone to center of sphere
```

```python
def F_ab(r_a, r_b, h): # a is the large one
    R_i = r_a/h
    R_j = r_b/h
    S = 1 + (1+R_j**2)/R_i**2
    return 0.5*(S-(S**2 - 4*(r_b/r_a)**2)**0.5) 
```

```python
def E_b(T):
    return 5.67e-8*T**4

def A_mat(R_12, R_13, R_14, R_23, R_24, R_34, R_2):
    A = np.zeros([5, 5])
    print(R_1)
    A[0,:] = np.array([-1/R_1 -1/R_12 - 1/R_13 - 1/R_14, 1/R_12, 1/R_13, 1/R_14, 1/R_1])
    A[1,:] = np.array([1/R_12, -1/R_2 - 1/R_12 - 1/R_23- 1/R_24, 1/R_23, 1/R_24, 0])
    A[2,:] = np.array([1/R_13, 1/R_23, -1/R_3 - 1/R_13 - 1/R_23 - 1/R_34, 1/R_34, 0])
    A[3,:] = np.array([1/R_14, 1/R_24, 1/R_34, -1/R_4 - 1/R_14 - 1/R_24 - 1/R_34,0])
    A[4,:] = np.array([0,0,1/R_3, 0,0])
    
    return A
    
   
def B_mat(q_inc, T_TC, T_DFT, T_inf):
    B = np.zeros([5, np.shape(T_TC)[0]])
    B[0,:] = 0
    B[1,:] = -E_b(T_TC)/R_2
    B[2,:] = -E_b(T_DFT)/R_3
    B[3,:] = -E_b(T_inf)/R_4
    B[4,:] = q_inc*A_3 + E_b(T_DFT)/R_3
    
    return B

```

```python
R_air = 287.05
P_atm = 101325

def density(P_atm, R_air, T): # T in K
    return P_atm/(R_air*T)
```

```python
def correlation_sphere(D,h,T,T_inf): # T in C
    
    T_eval = T_inf
    nu_inf = dft.air_props(T_eval).kinematic_viscosity(T_eval)
    nu_s = dft.air_props(T).kinematic_viscosity(T)
    alpha = dft.air_props(T_eval).thermal_diffusivity(T_eval)
    k = dft.air_props(T_eval).thermal_conductivity(T_eval)
    Pr = nu_inf/alpha
    Nu = h*D/k
    Re = np.zeros(T.shape[0])
    mu_inf = nu_inf*density(P_atm, R_air,T_inf+273)
    mu_s = nu_s*density(P_atm, R_air, T+273)
    mu_ratio = mu_inf/mu_s

    def f(x): 
        return 0.4*x**0.5 + 0.06*x**(2/3) - (Nu_eq-2)/(Pr**0.4*mu_ratio_eq**0.25)

    
    for idx in range(T.shape[0]):
        Nu_eq = Nu[idx]
        mu_ratio_eq = mu_ratio[idx]
        Re[idx], info = opt.brentq(f,-5000, 1000000, full_output=True)
        # this returns Re(TC) given h TC
        
    u_char = Re*nu_inf/D # get characteristic velocity
    return u_char
```

```python
def DFT_correlation(u, L_ch, T):
    T_eval = 0.5*(T + T_inf)
    nu = dft.air_props(T_eval).kinematic_viscosity(T_eval)
    alpha = dft.air_props(T_eval).thermal_diffusivity(T_eval)
    k = dft.air_props(T_eval).thermal_conductivity(T_eval)
    Pr = nu/alpha
    Re = u*L_ch/nu
    Nu = 0.664*Re**0.5*Pr**(1/3)
    return k*Nu/L_ch # returns h
```
